```python
config = load_config('../configs/tum_config_notebooks.yaml')
data_config = config['data']
data_config['name'] = data_config['name'][0]
```

```python
dataset = TUMDataset(data_config)
```

```python
rows = 3

for i in range(rows):
    
    index = np.random.randint(low=0, high=dataset.__len__())
    item = dataset.__getitem__(index)
    
    frames = []
    titles = []
    
    title_image = "{}; {}".format(item[NAME], item[IMAGE].shape)
    title_depth = "{}".format(item[DEPTH].shape)
    
    frames.append(item[IMAGE].transpose((1, 2, 0)))
    frames.append(grayscale2rgb(item[DEPTH]).transpose((1, 2, 0)))
    
    titles.append(title_image)
    titles.append(title_depth)
    
    plot_images(frames, titles, dpi=200)
    
```
